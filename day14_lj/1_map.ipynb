# 折线颜色和形状设置

```python
#生成一个红色虚线的折线图案列表
# 导入库
import matplotlib.pyplot as plt
import numpy as np
# 生成数据
x = np.linspace(0, 10, 100)#linspace函数用于生成一个等差数列，参数分别为起始值，终止值，元素个数
y = np.sin(x)    # 生成正弦曲线
# 绘制折线图
plt.plot(x, y, color='red', linestyle='--')    # 绘制红色虚线
# 显示图形
plt.show()    # 显示图形
```

# 折点样式

```python
#生成一个圆圈的折点样式案例
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']#rcParams是一个字典，用于设置matplotlib的全局参数，font.sans-serif是一个字体列表，用于设置字体
mpl.rcParams['axes.unicode_minus'] = False#axes.unicode_minus是一个布尔值，用于设置负号的显示
# 生成数据
x = np.linspace(0, 2*np.pi, 100)#linspace是一个函数，用于生成一个等差数列，第一个参数是起始值，第二个参数是终止值，第三个参数是元素个数
y = np.sin(x)   # 生成正弦曲线
# 绘制折线图
plt.plot(x, y, marker='o', markersize=10, markerfacecolor='red', markeredgecolor='blue', markeredgewidth=2)  # 标记点为圆形
plt.title('折线图')
plt.xlabel('X轴')
plt.ylabel('Y轴')   # 标记点为圆形  
```

# 绘制X轴和y轴的刻度

```python
from matplotlib import pyplot as plt
x = range(2,26,2) # x轴的位置
y = [random.randint(15, 30) for i in x]
plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
x_ticks = ["{}:00".format(i) for i in x]#列表生成式,将x轴的位置转换为时间
plt.xticks(x,x_ticks,rotation=45) # rotation 旋转
y_ticks = ["{}℃".format(i) for i in range(min(y),max(y)+1)]
plt.yticks(range(min(y),max(y)+1),y_ticks)
plt.grid(alpha=0.4) # alpha 透明度
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("10点到12点的气温变化")
plt.plot(x,y)
plt.savefig("./t3.png") # 保存图片
```

# 一线多图

```python
#完成一个一线多图的案例
# 导入模块
import matplotlib.pyplot as plt
import numpy as np
# 定义x，y
x = np.linspace(-3,3,50)# 从-3到3，50个点
y1 = 2*x+1
y2 = x**2
# 定义画布
plt.figure()    # 定义一个图像窗口
# 定义第一个子图
plt.subplot(2,2,1)    # 一行两列第一个
plt.plot(x,y1)    # 画第一个图
# 定义第二个子图
plt.subplot(2,2,2)    # 一行两列第二个
plt.plot(x,y2)    # 画第二个图
#定义第三个子图
plt.subplot(2,2,3)    # 两行一列第一个
plt.plot(x,np.sin(x))    # 画第一个图

plt.subplot(2,2,4)    # 两行一列第二个
plt.plot(x,np.cos(x))    # 画第二个图
# 显示
plt.show()    # 显示图像窗口
```

# 一图多个坐标系子图

```python
#一图多个坐标系
import matplotlib.pyplot as plt
import numpy as np
#创建一个图
fig = plt.figure()      #创建一个图
#创建第一个坐标系
ax1 = fig.add_subplot(2,1,1)        #两行一列第一个
ax1.plot([1,2,3],[3,2,1])
#创建第二个坐标系
ax2 = fig.add_subplot(2,3,4)        #两行三列第四个
ax2.plot([1,2,3],[1,2,3])
#创建第三个坐标系
ax3 = fig.add_subplot(2,3,5)        #两行三列第五个
ax3.plot([1,2,3],[3,1,2])
#创建第四个坐标系
ax4 = fig.add_subplot(2,3,6)        #两行三列第六个
ax4.plot([1,2,3],[2,3,1])
plt.show()
```

# 绘制条形图

```python
import matplotlib.pyplot as plt#pyplot 是matplotlib的一个模块，用于绘制图形
from matplotlib import font_manager#font_manager 管理字体
a = ['流浪地球','疯狂的外星人','飞驰人生','大黄蜂','熊出没·原始时代','新喜剧之王']
b = [38.13,19.85,14.89,11.36,6.47,5.93]
width = 0.5#条形图宽度
rects=plt.bar(range(len(a)),b,width,color='b')#绘制条形图，range(len(a))是条形图的横坐标，b是条形图的纵坐标，width是条形图的宽度，color是条形图的颜色
plt.xticks(range(len(a)),a,rotation=45)#设置条形图的横坐标，range(len(a))是条形图的横坐标，a是条形图的横坐标标签
for rect in rects:#设置条形图的高度标签
    height = rect.get_height()#获取条形图的高度
    plt.text(rect.get_x()+rect.get_width()/2,1.03*height,str(height),ha='center',va='bottom')#设置条形图的高度标签，rect.get_x()+rect.get_width()/2是条形图的横坐标，1.03*height是条形图的纵坐标，str(height)是条形图的高度标签，ha='center'是条形图的高度标签的水平对齐方式，va='bottom'是条形图的高度标签的垂直对齐方式
plt.show()
```

# 饼形图

```python
#做一个饼形图的案例
import matplotlib.pyplot as plt
#准备数据
labels = ['Frogs','Hogs','Dogs','Logs']
sizes = [15,30,45,10]
explode = (0,0.1,0.2,0) #突出显示
#开始画图
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90,rotatelabels=True)
plt.axis('equal') #显示为圆形
plt.show()
```
