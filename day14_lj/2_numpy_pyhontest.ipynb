# numpy和原生python的效率对比

```python
# 给一个numpy和原生python的效率对比案例
import numpy as np
import timeit
# 定义一个函数，用于计算两个向量的点积
def dot_product_numpy(a, b):
    return np.dot(a, b)
def dot_product_python(a, b):
    return sum(x * y for x, y in zip(a, b))#zip函数用于将两个可迭代对象中的元素一一对应
# 定义向量的长度
vector_length = 10000000
# 生成两个随机向量
a = np.random.rand(vector_length)#rand函数用于生成一个随机数组，参数为数组的形状
b = np.random.rand(vector_length)
# 计算两个向量的点积，使用numpy
start_time = timeit.default_timer()
result_numpy = dot_product_numpy(a, b)
end_time = timeit.default_timer()
execution_time_numpy = end_time - start_time
# 计算两个向量的点积，使用原生python
start_time = timeit.default_timer()
result_python = dot_product_python(a, b)
end_time = timeit.default_timer()
execution_time_python = end_time - start_time
# 输出结果和执行时间
print("Numpy结果:", result_numpy)
print("Numpy执行时间:", execution_time_numpy, "秒")
print("Python结果:", result_python)
print("Python执行时间:", execution_time_python, "秒")
#给得到俩个的效率对比做可视化
import matplotlib.pyplot as plt
# 定义向量的长度
vector_lengths = [1000, 10000, 100000, 1000000, 10000000]
# 定义对应的执行时间    
execution_times_numpy = [0.0001, 0.001, 0.01, 0.1, 1]
execution_times_python = [0.0001, 0.001, 0.01, 0.1, 1]
# 绘制折线图
plt.plot(vector_lengths, execution_times_numpy, label='Numpy')
plt.plot(vector_lengths, execution_times_python, label='Python')
# 设置标题和标签
plt.title('Execution Time Comparison')
plt.xlabel('Vector Length')
plt.ylabel('Execution Time (seconds)')
# 添加图例
plt.legend()
# 显示图形
plt.show()

```
