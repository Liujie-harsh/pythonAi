```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("world food production.csv",encoding="utf-8")
df.head()
```

```python
data=df.dropna()#删除缺失值
data.info()
```

```python

# 绘制柱状图进行可视化
plt.figure(figsize=(10, 6))
# 创建一个新的列，将年份分组为每五年一组
data['Year_Group'] = (data['Year'] // 5) * 5
# 统计每五年组的数据数量
year_group_counts = data['Year_Group'].value_counts().sort_index()
# 用新的统计结果替换原有的统计结果
year_counts = year_group_counts

year_counts.plot(kind='bar')
# 引入 matplotlib 支持汉字显示的配置
plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

plt.title('按年份统计的数据分布')
plt.xlabel('年份')
plt.ylabel('数量')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
