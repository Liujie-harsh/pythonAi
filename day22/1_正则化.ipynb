```python
# 导入必要的库
import numpy as np  # 用于数值计算
import matplotlib.pyplot as plt  # 用于数据可视化
from sklearn.datasets import make_regression  # 用于生成回归问题的模拟数据

# 生成模拟数据
# n_samples: 样本数量为100
# n_features: 特征数量为1（一元线性回归）
# noise: 添加噪声强度为10
# random_state: 设置随机种子，确保结果可复现
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)
X = X.reshape(-1, 1)  # 将X转换为二维数组，形状为(100, 1)

# 添加偏置项（截距项）
# np.c_是按列连接两个矩阵的函数
# 添加一列1，使得X_b的形状变为(100, 2)，第一列全为1，用于学习截距
X_b = np.c_[np.ones((100, 1)), X]

# 定义梯度下降函数
def gradient_descent(X, y, theta, learning_rate=0.01, n_iterations=1000):
    """
    实现批量梯度下降算法
    参数:
        X: 特征矩阵，包含偏置项
        y: 目标值
        theta: 初始参数向量
        learning_rate: 学习率，控制参数更新步长
        n_iterations: 迭代次数
    返回:
        theta: 最终的参数值
        cost_history: 记录每次迭代的损失值
    """
    m = len(y)  # 样本数量
    cost_history = []  # 用于记录每次迭代的损失值
    
    for i in range(n_iterations):
        # 计算预测值
        prediction = np.dot(X, theta)
        # 计算误差
        error = prediction - y
        # 计算梯度（损失函数对参数的导数）
        gradients = 2/m * X.T.dot(error)
        # 更新参数
        theta = theta - learning_rate * gradients
        # 计算当前损失值（均方误差）
        cost = np.mean(error ** 2)
        cost_history.append(cost)
    
    return theta, cost_history

# 初始化模型参数
# 随机初始化两个参数（截距和斜率）
theta = np.random.randn(2, 1)

# 训练模型
theta_final, cost_history = gradient_descent(X_b, y, theta)#gradient_descent梯度下降

# 创建一个12x5的图形
plt.figure(figsize=(12, 5))
# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号


# 绘制数据点和拟合线（左图）
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='数据点')  # 绘制原始数据点
plt.xlabel('X')
plt.ylabel('y')
plt.title('线性回归拟合结果')
plt.legend()

# 绘制损失函数随迭代次数的变化（右图）
plt.subplot(1, 2, 2)
plt.plot(cost_history)
plt.xlabel('迭代次数')
plt.ylabel('损失值')
plt.title('损失函数收敛过程')

# 调整子图布局
plt.tight_layout()
plt.show()

# 打印最终得到的模型参数（截距和斜率）
print("最终参数: ", theta_final.flatten())
```

```python
# Lasso回归示例 - L1正则化的线性回归
# 导入必要的库
import numpy as np  # 用于数值计算
import matplotlib.pyplot as plt  # 用于数据可视化
from sklearn.linear_model import Lasso  # 导入Lasso回归模型
from sklearn.preprocessing import StandardScaler  # 用于特征标准化

# 生成示例数据
np.random.seed(42)  # 设置随机种子，确保结果可复现
# 生成100个样本，特征值在0到2之间均匀分布
X = 2 * np.random.rand(100, 1)
# 生成目标值：y = 4 + 3x + 噪声
# 这里4是真实的截距，3是真实的系数，randn添加正态分布噪声
y = 4 + 3 * X + np.random.randn(100, 1)

# 特征标准化
# 对特征进行标准化处理（均值为0，方差为1）
# 这是很重要的预处理步骤，因为Lasso对特征的尺度很敏感
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 创建和训练Lasso模型
# alpha参数控制L1正则化的强度：
# - alpha越大，正则化越强，模型越简单，系数越稀疏
# - alpha越小，正则化越弱，模型越复杂，可能过拟合
lasso = Lasso(alpha=0.1)
lasso.fit(X_scaled, y)  # 训练模型

# 生成预测数据
# 创建均匀分布的测试点，范围从-2到2
X_new = np.linspace(-2, 2, 100).reshape(100, 1)
# 对测试数据进行标准化（使用与训练数据相同的缩放器）
X_new_scaled = scaler.transform(X_new)
# 使用训练好的模型进行预测
y_pred = lasso.predict(X_new_scaled)

# 设置图形参数
plt.figure(figsize=(12, 5))  # 设置图形大小
# 设置中文字体，避免中文显示乱码
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 绘制第一个子图：数据拟合结果
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='数据点')  # 绘制原始数据散点图
plt.plot(X_new, y_pred, color='red', label='Lasso回归')  # 绘制拟合曲线
plt.xlabel('X')
plt.ylabel('y')
plt.title('Lasso回归拟合结果')
plt.legend()

# 研究不同正则化强度（alpha值）对模型的影响
# 测试一系列不同的alpha值，从非常小（0.0001）到很大（10）
alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10]
coefs = []  # 存储不同alpha值对应的模型系数
for alpha in alphas:
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_scaled, y)
    coefs.append(lasso.coef_[0])  # 保存每个alpha值对应的系数

# 绘制第二个子图：正则化强度与系数的关系
plt.subplot(1, 2, 2)
# 使用对数刻度绘制alpha值（x轴）与系数（y轴）的关系
plt.semilogx(alphas, coefs, marker='o')
plt.xlabel('正则化系数 (alpha)')
plt.ylabel('模型系数')
plt.title('Lasso系数随正则化强度变化')

# 调整子图布局，避免重叠
plt.tight_layout()
plt.show()

# 打印最终模型的参数
print("Lasso回归系数:", lasso.coef_[0])  # 打印特征的系数
print("Lasso回归截距:", lasso.intercept_)  # 打印模型的截距
```

```python
# 导入岭回归模型
from sklearn.linear_model import Ridge

# 创建并训练岭回归模型
ridge = Ridge(alpha=1.0)  # alpha是正则化强度参数
ridge.fit(X_scaled, y)

# 使用训练好的模型进行预测
y_pred_ridge = ridge.predict(X_new_scaled)

# 设置图形参数
plt.figure(figsize=(12, 5))
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


# 研究不同正则化强度的影响
alphas = [0.0001, 0.001, 0.01, 0.1, 1, 10]
ridge_coefs = []
for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_scaled, y)
    ridge_coefs.append(ridge.coef_[0])

# 绘制第二个子图：正则化强度与系数的关系
plt.subplot(1, 2, 2)
plt.semilogx(alphas, ridge_coefs, marker='o')
plt.xlabel('正则化系数 (alpha)')
plt.ylabel('模型系数')
plt.title('岭回归系数随正则化强度变化')

# 调整布局并显示图形
plt.tight_layout()
plt.show()


# 打印最终模型的参数
print("岭回归系数:", ridge.coef_[0])
print("岭回归截距:", ridge.intercept_)
```
